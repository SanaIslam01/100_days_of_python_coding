```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

titanic=sns.load_dataset("titanic")
titanic.head()
```

```python
sns.boxplot(x="survived",y="age",data=titanic)
```

```python
p1=sns.boxplot(x="survived",y="age",showmeans=True,data=titanic)
```

```python
p1=sns.boxplot(x="survived",y="age",showmeans=True,meanprops = {"marker":"*","markersize":"12","markeredgecolor":"red"},data=titanic)
```

```python
p1=sns.boxplot(x="survived",y="age",
               showmeans=True,meanprops = {"marker":"*","markersize":"12","markeredgecolor":"red"},
               data=titanic)
plt.xlabel("how many survived"),
plt.ylabel("age (years)"),
plt.title("box plot of kitny jeeye r kitny mary")

plt.show()
```

```python
p1=sns.boxplot(x="survived",y="age",
               showmeans=True,meanprops = {"marker":"*","markersize":"12","markeredgecolor":"red"},
               data=titanic)
plt.xlabel("how many survived",size=12),
plt.ylabel("age (years)",size=14),
plt.title("box plot of kitny jeeye r kitny mary", size=16,weight="bold")

plt.show()
```
