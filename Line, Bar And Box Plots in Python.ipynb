### How to add Title in a Plot

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset

phool = sns.load_dataset("iris")
phool

# draw a line plot
sns.lineplot(x="sepal_length",y="sepal_width",data=phool)
# Adding title
plt.title("naba its your time")
# showing plot
plt.show()
```

### Adding x and y Limits

#### how to add x Limits

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset

phool = sns.load_dataset("iris")
phool

# draw a line plot
sns.lineplot(x="sepal_length",y="sepal_width",data=phool)
# Adding title
plt.title("naba its your time")
# Adding x limit
plt.xlim(2)
# Showing plot
plt.show()
```

#### How to add y limit

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset

phool = sns.load_dataset("iris")
phool

# draw a line plot
sns.lineplot(x="sepal_length",y="sepal_width",data=phool)
# Adding title
plt.title("naba its your time")
# Adding y limit
plt.ylim(1)
# showing plot
plt.show()
```

### Set Styles

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt


# load dataset
phool = sns.load_dataset("iris")
phool

# draw a line plot
sns.lineplot(x="sepal_length",y="sepal_width",data=phool)
# Adding title
plt.title("naba its your time")
# Adding style or setting style
sns.set_style("dark")
sns.set_style(style=None, rc=None)
# showing plot
plt.show()
```

### Size of Figure 

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
phool = sns.load_dataset("iris")

# change figure size 
plt.figure(figsize=(8,6))

# draw a line plot
sns.lineplot(x="sepal_length",y="sepal_width",data=phool)
# Adding title
plt.title("naba its your time")
# Adding style or setting style
sns.set_style("dark")
# showing plot
plt.show()
```

### Adding hue .... here hue means color

### Adding order

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
kashti = sns.load_dataset("titanic")
kashti

# draw a bar plot
sns.barplot(x="who",y="alone",hue="sex",data=kashti,order=["alone","who"])
# showing bar plot
plt.show()
```

### Adding confidence interval

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
kashti = sns.load_dataset("titanic")
kashti

# draw a bar plot
sns.barplot(x="who",y="alone",hue="sex",data=kashti,color="gray", errorbar=None,
           palette='pastel')
# showing bar plot
plt.show()
```

### Adding Estimator

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt
from numpy import mean
# load dataset
kashti = sns.load_dataset("titanic")
kashti

# draw a bar plot
sns.barplot(x="who",y="alone",hue="fare",data=kashti,color="gray", estimator=mean,
           palette='pastel')
# showing bar plot
plt.show()
```

### How to Add saturation

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt
from numpy import mean
# load dataset
kashti = sns.load_dataset("titanic")
kashti

# draw a bar plot
sns.barplot(x="who",y="alone",hue="fare",data=kashti,color="gray",estimator=mean,palette='pastel',saturation=0.2)
# showing bar plot
plt.show()
```

### Horizontal graph

### to make like a border graph

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
kashti = sns.load_dataset("titanic")
kashti

# draw a bar plot
sns.barplot(x="class",y="fare",hue="sex",data=kashti,linewidth=5,facecolor=(1,1,1,0),errcolor=".2",edgecolor=".2")
# showing bar plot
plt.show()
```
